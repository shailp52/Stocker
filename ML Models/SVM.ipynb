```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from ta import *

data = pd.read_csv('SENSEX.csv')
data = data.drop('Close',1)
data = data.rename(columns={'Adj Close':'Close'})
data.head()
```

```python
X = data.iloc[:,1:]
y = np.where (data['Close'].shift(-1) > data['Close'],1,-1)
X = X.drop('Open',1)
X.head()
```

```python
#Calculating the Technical Indicators
X['SMA'] = X['Close'].rolling(window=10).mean()
X['EMA'] = X['Close'].ewm(span=20, adjust=False).mean()
X['STCK'] = momentum.stoch(X['High'], X['Low'], X['Close'], n=14, fillna=False)
X['STCD'] = momentum.stoch_signal(X['High'], X['Low'], X['Close'], n=14, d_n=3, fillna=False)
X['MACD'] = trend.macd(X['Close'], n_fast=12, n_slow=26, fillna=False)
X['RSI'] = momentum.rsi(X['Close'], n=10,fillna=False)
X['WR'] = momentum.wr(X['High'], X['Low'], X['Close'], lbp=14, fillna=False)
X['AD'] = volume.acc_dist_index(X['High'], X['Low'], X['Close'], X['Volume'], fillna=False)
X['CCI'] = trend.cci(X['High'], X['Low'], X['Close'], n=20, c=0.015, fillna=False)
X = X.iloc[25:,:]
X.head(440)
```

```python
# new_X = ['SMA','EMA','STCK','STCD','MACD','RSI','WR','AD','CCI']
X.size
```

```python
k=0
new_X = []
j = []
X_list = [X.values.tolist()]
len(X_list[0])

# print(X_list[0])

for i in X_list[0]:
    l = []
#     print(i)
#     j=i
#     print(j)
    if k==0:
        k=1
        j=i
        continue
    
    #SMA
    if i[2]>=i[4]:
        l.append(1)
    else:
        l.append(-1)
#     print(i[2])
    
    #EMA
    if i[2]>=i[5]:
        l.append(1)
    else:
        l.append(-1)
#     print((i[2]>=i[5]))

    #Stck
    if i[6] >= j[6]:
        l.append(1)
    else:
        l.append(-1)
#     print(l)
           
    #STCD
    if i[7] >= j[7]:
        l.append(1)
    else:
        l.append(-1)
#     print(l)
    
    #MACD
    if i[8] >= j[8]:
        l.append(1)
    else:
        l.append(-1)
#     print(l)
        
    #RSI
    if i[9]<30:
        l.append(1)
    elif i[9]>70:
        l.append(-1)
    else:
        if i[9]>=j[9]:
            l.append(1)
        else:
            l.append(-1)
#     print(l)
        
    #WR
    if i[10] >= j[10]:
        l.append(1)
    else:
        l.append(-1)
#     print(l)

    #AD
    if i[11] >= j[11]:
        l.append(1)
    else:
        l.append(-1)
        
    #CCI
    if i[12]<-200:
        l.append(1)
    elif i[12]>200:
        l.append(-1)
    else:
        if i[12]>=j[12]:
            l.append(1)
        else:
            l.append(-1)
#     print(i[11])
#     print(j[11])
    

#     print(l)
    
        
    #Append to new_X
    new_X.append(l)
    j=i
```

```python
new_X = pd.DataFrame(new_X, columns =['SMA','EMA','STCK','STCD','MACD','RSI','WR','AD','CCI'], dtype = int)
new_X.head()
```

```python
from sklearn.model_selection import KFold
from sklearn.preprocessing import MinMaxScaler
from sklearn import svm


svm_model = svm.SVC(kernel='poly', C=1e5, degree=1)
cv = KFold(n_splits=10, random_state=42, shuffle=False)

scaler = MinMaxScaler(feature_range=(-1, 1))
new_X = scaler.fit_transform(new_X)

scores = []
for train_index, test_index in cv.split(new_X):
    X_train, X_test, y_train, y_test = new_X[train_index], new_X[test_index], y[train_index], y[test_index]
    svm_model.fit(X_train, y_train)
    scores.append(svm_model.score(X_test, y_test))
```
